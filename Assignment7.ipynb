```python
#percent between 1 and 100
import random
def simulation(percent):
    ans = []
    i = 0
    while i < 1000:
        side = random.randint(1, 2)
        #heads
        if side == 1:
            truth = random.randint(1, 100)
            if truth <= percent:
                ans.append("yes")
            else:
                ans.append("no")
        #tails
        else:
            side2 = random.randint(1, 2)
            #heads
            if side2 == 1:
                ans.append("yes")
            #tails
            else:
                ans.append("no")
        i += 1
    return ans
```

```python
def run100(percent):
    i = 0
    arr = []
    while i < 100:
        arr.append(simulation(percent))
        i += 1
    return arr

#print(run100(50))
#print(len(run100(50)))
```

```python
fifty = run100(50)
twentyfive = run100(25)
ten = run100(10)
one = run100(1)
```

```python
#takes in array of arrays
def propyes(arr):
    ans = []
    for lst in arr:
        y = 0
        n = 0
        for element in lst:
            if element == 'yes':
                y += 1
            else:
                n += 1
        total = y + n
        ans.append(y / total)
    return ans
            

#propyes(fifty)
#propyes(twentyfive)
#propyes(ten)
#propyes(one)
```

```python
#coin probability deciding to not answer truthfully is 1/8, and 1/4, and 3/4
def simulation_bias(percent, bias):
    ans = []
    i = 0
    bias = bias * 100
    while i < 1000:
        side = random.randint(1, 100)
        #heads
        if side > bias:
            truth = random.randint(1, 100)
            if truth <= percent:
                ans.append("yes")
            else:
                ans.append("no")
        #tails
        else:
            side2 = random.randint(1, 2)
            #heads
            if side2 == 1:
                ans.append("yes")
            #tails
            else:
                ans.append("no")
        i += 1
    return ans
```

```python
def run100_bias(percent, bias):
    i = 0
    arr = []
    while i < 100:
        arr.append(simulation_bias(percent, bias))
        i += 1
    return arr
```

```python
from scipy.stats import laplace  
def propyes_eps(arr, epsilon):
    temparr = propyes(arr)
    ret = []
    scl = 1/epsilon
    for prop in temparr:
        #print(prop)
        #print(lp)
        ret.append(prop + laplace.rvs(scale = scl))
    return ret
```

```python
import statistics

age = income_data['age']
maxage = max(age)
minage = min(age)
mean = statistics.mean(age)

print('Max: ')
print(maxage)
print('Min:')
print(minage)
print('Mean:')
print(mean)
```

```python
#chosing range from 17 to 59 -> same distance on either side of the mean
#arr should be the age array
#bounds are inclusive
def bounding(arr, minval, maxval):
    ret = []
    for age in arr:
        if age <= maxval and age >= minval:
            ret.append(age)
    return ret
    
bounded = bounding(age, 17, 59)
#print(bounded)
```

```python
#creating and calling function for differentially private information
def dp(epsilon, lst):
    scl = 1/epsilon
    ret = []
    for a in lst:
        #print(type(a))
        ret.append(a + laplace.rvs(scale = scl))
    return ret

diff_priv = dp(0.1, bounded)
#print(diff_priv)
not_diff_priv = bounded
```

```python
def difference(arr1, arr2):
    i = 0
    ret = []
    while i  < len(arr1):
        ret.append(abs(arr1[i] - arr2[i]))
        i += 1
    return ret

difference = difference(diff_priv,not_diff_priv)
#print(difference)
```

```python
import matplotlib.pyplot as plt

plt.hist(difference)
plt.title("Average Difference Between the True Answer and the Differentially Private Answers ")
plt.xlabel("Average Difference")
plt.ylabel("Frequency")
plt.show()
```
